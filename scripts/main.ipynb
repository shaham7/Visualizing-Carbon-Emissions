```python
import pandas as pd
from pathlib import Path

data_dir = Path('..') / 'data'
emissions = pd.read_csv(data_dir / 'emissions.csv')
renewables = pd.read_csv(data_dir / 'renewables.csv')
gdp = pd.read_csv(data_dir / 'GDP.csv')

renewables
```

```python
renewables_long = renewables.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
                                  var_name='Year', value_name='Renewable Energy %')
try:
  renewables_long['Year'] = renewables_long['Year'].str.extract('YR(\d{4})').astype(int)
except (TypeError, ValueError):
  print("Warning: Error converting 'Year' column. Check data format.")
  # Optionally: You can choose to drop rows here or fill missing years

# You can now choose to drop rows with missing years:
renewables_long = renewables_long.dropna(subset=['Year'])  # Drop rows where year is NaN

renewables_long
```
